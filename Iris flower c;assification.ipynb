```python
#simporting libraries used for encoding
from sklearn.preprocessing import LabelEncoder
from pandas.core.dtypes.common import is_numeric_dtype
le = LabelEncoder()
```

```python
for i in df.columns:
    if is_numeric_dtype(df[i]):
        continue
    df[i] = le.fit_transform(df[i])
```
